(2) Jelaskan apa yang dimaksud dengan unsupervised learning

Unsupervised Learning adalah salah satu jenis pembelajaran mesin di mana model dilatih menggunakan data yang tidak memiliki label atau output target. Tujuan utama unsupervised learning adalah untuk menemukan pola tersembunyi atau struktur dalam data. Beberapa konsep penting dalam unsupervised learning:

* Clustering: Mengelompokkan data ke dalam beberapa kelompok berdasarkan kesamaan fitur. Contoh algoritma clustering adalah K-Means, Hierarchical Clustering, dan DBSCAN.

* Dimensionality Reduction: Mengurangi jumlah fitur dalam data tanpa kehilangan informasi penting. Contoh algoritma dimensionality reduction adalah PCA (Principal Component Analysis) dan t-SNE (t-distributed Stochastic Neighbor Embedding).

* Anomaly Detection: Mendeteksi data yang berbeda secara signifikan dari mayoritas data. Ini digunakan dalam deteksi penipuan, diagnosis penyakit, dan keamanan jaringan.

```python
!pip install requests wordcloud nltk matplotlib seaborn

import requests


url = 'https://shared.djambred.my.id/uas-nim-ganjil.txt'


response = requests.get(url)
text = response.text


print(text[:1000])  
```

```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

nltk.download('punkt')
nltk.download('stopwords')


def clean_text(text):

    text = text.lower()
    
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    text = re.sub(r'\d+', '', text)
    return text

cleaned_text = clean_text(text)


tokens = word_tokenize(cleaned_text)


stop_words = set(stopwords.words('indonesian'))
filtered_tokens = [word for word in tokens if word not in stop_words]


print(filtered_tokens[:10])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_tokens))


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from collections import Counter
import seaborn as sns


word_freq = Counter(filtered_tokens)


most_common_words = word_freq.most_common(20)


words, counts = zip(*most_common_words)


plt.figure(figsize=(10, 5))
sns.barplot(x=list(words), y=list(counts))
plt.xticks(rotation=45)
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.title('20 Kata Paling Sering Digunakan')
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')


sid = SentimentIntensityAnalyzer()


sentiment_scores = sid.polarity_scores(cleaned_text)


print(sentiment_scores)


labels = ['Negatif', 'Netral', 'Positif']
sizes = [sentiment_scores['neg'], sentiment_scores['neu'], sentiment_scores['pos']]
colors = ['red', 'gold', 'green']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribusi Sentimen')
plt.show()
```
